```python
import speech_recognition as sr

# Initialize the recognizer
recognizer = sr.Recognizer()

# Load the audio file
audio_file = 'converted_audio1.wav'  # replace with your uploaded file name

# Use SpeechRecognition to open the file
with sr.AudioFile(audio_file) as source:
    audio = recognizer.record(source)  # read the entire audio file

# Recognize speech using Google's API
try:
    text = recognizer.recognize_sphinx(audio)
    print("Transcription: " + text)
except sr.UnknownValueError:
    print("Google Speech Recognition could not understand the audio.")
except sr.RequestError as e:
    print(f"Could not request results from Google Speech Recognition service; {e}")
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
model_name = "NousResearch/Llama-2-7b-chat-hf"
prompt = "Tell me about the Parkinson's disease and the stages of those, also tell me the precautions that I can take when I know a person with parkinson's disease and how to understand if a person is advancing to next stages"




model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", load_in_4bit=True)
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)


```

```python
model_inputs = tokenizer(prompt, return_tensors="pt").to("cuda:0")
output = model.generate(**model_inputs,max_length=1000)
print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
import re

# ... (rest of your code)

# Split the conversation into individual messages based on timestamps
messages = re.split(r"\d{2}:\d{2}:\d{2}", cleaned_text)

# Strip whitespace and filter out any empty messages
messages = [message.strip() for message in messages if message.strip()]

# Combine messages into a single input string
combined_messages = " ".join(messages)
prompt="Make a summary of the following content"
final= prompt + combined_messages

# Tokenize the combined messages
model_inputs = tokenizer(final, return_tensors="pt", padding=True, truncation=True).to("cuda:0")

# Generate summary
output = model.generate(**model_inputs, max_length=2000)

# Decode summary
summary = tokenizer.decode(output[0], skip_special_tokens=True)

summary_text = summary.split("Summary:")[-1].strip()

print(summary_text)
```
